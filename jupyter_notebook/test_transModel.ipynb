```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="pysentimiento/bert-pt-emotion")
```

```python
text = "Estou muito feliz hoje!"
result = pipe(text)
print(result)
```
