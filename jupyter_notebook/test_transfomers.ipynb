```python
# Use a pipeline as a high-level helper
from transformers import pipeline
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
```

```python
frases_emocoes_en = [
    "Anxious about tomorrow, can you recommend something relaxing to listen to before bed?",
    "I'm kind of sad, I'd like to hear something calmer, like a melancholic ballad.",
    "I'm very happy today, I want an upbeat song to celebrate.",
    "I just passed a difficult exam, I need a soundtrack to celebrate!",
    "I ended a relationship, I don't want anything too lively right now.",
    "It's raining and I just wanted a calm playlist to stay at home.",
    "I'm really angry, I need some heavy music to blow off steam.",
    "I'm nostalgic today, recommend me some old songs from the 90s.",
    "Anxious and unfocused, I need something instrumental to calm me down.",
    "I'm euphoric, I want very energetic songs to keep the energy up!",
    "I'm in love, I want romantic songs to set the mood.",
]
```

```python
for frase in frases_emocoes_en:
    resultado = classifier(frase)
    print(f"Frase: {frase}")
    print("Resultado:", resultado)
    print()
```
